# Checking for Mixed Type Columns

```python
# Checking df_game for mixed type columns
for col in df_fifa20.columns.tolist():
  weird = (df_fifa20[[col]].applymap(type) != df_fifa20[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_fifa20[weird]) > 0:
    print (col)
    print (col + ' has mixed type data')
  else:
    print (col + ' data type is uniform')
```

```python
# Looking at the rows in df_nan, it quickly becomes clear that it isn’t necessary to do imputation
# creating a new data frame and filter out the ones that aren’t missing into a subset dataframe and
# continue wih the analysis with this new dataframe.
```

```python
# Creating new data frame using 'contract_valid_until'
df_fifa20_clean = df_fifa20[df_fifa20['contract_valid_until'].isnull() == False]
```

```python
# Checking df_fifa20_clean
df_fifa20_clean.isnull().sum()
```

```python
# imputing the mean values in the df_fifa20_clean dataframe where there is missing rows 
```

```python
#Replacing the release_clause_eur column with the average
df_fifa20_clean['release_clause_eur'].fillna(df_fifa20_clean['release_clause_eur'].mean(), inplace = True)
#Replacing the pace column with the average
df_fifa20_clean['pace'] = df_fifa20_clean['pace'].fillna(df_fifa20_clean['pace'].mean())
#Replacing the passing column with the average
df_fifa20_clean['passing'].fillna(df_fifa20_clean['passing'].mean(), inplace = True)
#Replacing the dribbling column with the average
df_fifa20_clean['dribbling'].fillna(df_fifa20_clean['dribbling'].mean(), inplace = True)
#Replacing the defending column with the average
df_fifa20_clean['defending'].fillna(df_fifa20_clean['defending'].mean(), inplace = True)
#Replacing the physic column with the average
df_fifa20_clean['physic'].fillna(df_fifa20_clean['physic'].mean(), inplace = True)
#Replacing the shooting column with the average
df_fifa20_clean['shooting'].mean()
# Imputing mean
df_fifa20_clean['shooting'].fillna(52, inplace=True)
```

```python
# Checking df_fifa20_clean
df_fifa20_clean.isnull().sum()
```

```python
# Checking for Uniqueness of the dataframe
df_fifa20_clean.ID.unique().shape
```

```python
# checking body_type Column
# Messi is not a body_type
df_fifa20_clean['body_type'].value_counts(dropna = False)
```

```python
# Changing body_type where there is an error
df_fifa20_clean['body_type'][df_fifa20_clean['body_type'] == 'Messi'] = 'Lean'
df_fifa20_clean['body_type'][df_fifa20_clean['body_type'] == 'C. Ronaldo'] = 'Normal'
df_fifa20_clean['body_type'][df_fifa20_clean['body_type'] == 'Neymar'] = 'Lean'
df_fifa20_clean['body_type'][df_fifa20_clean['body_type'] == 'PLAYER_BODY_TYPE_25'] = 'Normal'
df_fifa20_clean['body_type'][df_fifa20_clean['body_type'] == 'Courtois'] = 'Lean'
df_fifa20_clean['body_type'][df_fifa20_clean['body_type'] == 'Shaqiri'] = 'Stocky'
df_fifa20_clean['body_type'][df_fifa20_clean['body_type'] == 'Akinfenwa'] = 'Stocky'
```
